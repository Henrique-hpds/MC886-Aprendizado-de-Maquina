#### **Addressing the Vanishing or Exploding Gradients problem and Normalization**

Normalizing image data before feeding it to a neural network model is essential for stable and efficient training. Normalization scales pixel values to a consistent range or distribution, helping the model converge faster and preventing issues like vanishing or exploding gradients, which can derail training.

1. **Min-Max Normalization**  
   - **What it does**: Scales pixel values to a specific range, typically [0,1].  
   - **How it works**: For images with pixel values ranging from 0 to 255, this involves dividing each pixel by 255.

2. **Standardization (Z-score Normalization)**  
   - **What it does**: Adjusts pixel values to have a mean of 0 and a standard deviation of 1.  
   - **How it works**: Subtract the mean pixel value and divide by the standard deviation, calculated either across the entire dataset or per channel for color images.

3. **Per-Channel Normalization**  
   - **What it does**: Normalizes each channel (e.g., Red, Green, Blue in RGB images) separately, often using channel-specific mean and standard deviation values.  
   - **How it works**: Apply standardization or predefined values per channel (e.g., ImageNet’s mean `[0.485, 0.456, 0.406]` and std `[0.229, 0.224, 0.225]`).

</br>

Example with per-channel normalization. In PyTorch, you can apply it using:

   ```python
   from torchvision import transforms

   normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
   # Apply to your image tensor after converting it to a tensor
   image = normalize(image)
   ```

##### **Why Normalize Images?**

Normalization is crucial for several reasons:

- **Stable Training**: Without normalization, large pixel values (e.g., 0–255) can lead to extreme weight updates, causing gradient issues like vanishing or exploding gradients that make training unstable or impossible.  
- **Faster Convergence**: By scaling inputs to a consistent range or distribution, the model can learn more efficiently, reducing training time.  
- **Improved Generalization**: Normalization ensures the model focuses on patterns in the data rather than being biased by pixel magnitude, leading to better performance on unseen data.

The vanishing or exploding gradients problem is not related only to data normalization. The model architecture can lead to this type of problem as well. For this task, check for techniques to mitigate this type of problem.

### Augumentation

````python
import torchvision.transforms as transforms

class AugRotate:
    def __init__(self):
        self.transforms = transforms.Compose([
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.5),
            transforms.RandomRotation(degrees=(-15, 15))
        ])

    def __call__(self, img, label):
        seed = random.randint(0, 99999)
        random.seed(seed)
        img = self.transforms(img)
        random.seed(seed)
        label = self.transforms(label)
        return img, label

class AugGaussianNoise:
    def __init__(self, mean_range=(0.0, 0.1), std_range=(0.01, 0.1)):
        self.mean_range = mean_range
        self.std_range = std_range

    def __call__(self, img):
        if isinstance(img, Image.Image):
            img = transforms.ToTensor()(img)

        mean = random.uniform(*self.mean_range)
        std = random.uniform(*self.std_range)
        noise = torch.randn_like(img) * std + mean
        noisy_img = img + noise
        noisy_img = torch.clamp(noisy_img, 0.0, 1.0)
        return transforms.ToPILImage()(noisy_img)

def load_tif_as_PIL(path):
    array = tiff.imread(path)

    if array.dtype != np.uint8:
        array = (255 * (array - array.min()) / (np.ptp(array) + 1e-8)).astype(np.uint8)

    return Image.fromarray(array)

def augmentate(image_dir, mask_dir):
    i = 0
    for image_path in sorted(glob.glob(os.path.join(image_dir, "*.tif"))):
        base = os.path.basename(image_path).replace('.tif', '')
        mask_path = os.path.join(mask_dir, base + '.png')

        img = load_tif_as_PIL(image_path).convert("L")
        label = Image.open(mask_path).convert("L")

        rotate = AugRotate()
        gauss = AugGaussianNoise()

        img_aug, label_aug = rotate(img, label)
        img_aug.save(f"{image_dir}/aug_rot_{i}.tif")
        label_aug.save(f"{mask_dir}/aug_rot_{i}.png")

        img_aug = gauss(img_aug)
        img_aug.save(f"{image_dir}/aug_rot_gauss_{i}.tif")
        label_aug.save(f"{mask_dir}/aug_rot_gauss_{i}.png")

        img_aug = gauss(img)
        img_aug.save(f"{image_dir}/aug_gauss_{i}.tif")
        label_aug.save(f"{mask_dir}/aug_gauss_{i}.png")

        i += 1

base_path = "/content/drive/My Drive/MC886/assignment_3/data/f3"

augmentate(f"{base_path}/images/train", f"{base_path}/annotations/train")
augmentate(f"{base_path}/images/val", f"{base_path}/annotations/val")
augmentate(f"{base_path}/images/test", f"{base_path}/annotations/test")
print("Dados augumentados com sucesso!")
````

````python
def generate_csv(image_dir, mask_dir, output_csv_path, num_classes=6, threshold=0.1):
    rows = []
    for image_path in sorted(glob.glob(os.path.join(image_dir, "*.tif"))):
        base = os.path.basename(image_path).replace('.tif', '')
        mask_path = os.path.join(mask_dir, base + '.png')

        mask = Image.open(mask_path)
        mask_np = np.array(mask)
        total_pixels = mask_np.size

        labels = np.zeros(num_classes, dtype=int)
        classes, counts = np.unique(mask, return_counts=True)

        for classe, count in zip(classes, counts):
            proportion = count / total_pixels
            if proportion >= threshold:
                labels[classe] = 1

        rows.append([image_path, labels.tolist()])

    df = pd.DataFrame(rows, columns=["image_path", "label"])
    df.to_csv(output_csv_path, index=False)


base_path = "/content/drive/My Drive/MC886/assignment_3/data/f3"

generate_csv(f"{base_path}/images/train", f"{base_path}/annotations/train", f"{base_path}/train.csv")
generate_csv(f"{base_path}/images/val", f"{base_path}/annotations/val", f"{base_path}/val.csv")
generate_csv(f"{base_path}/images/test", f"{base_path}/annotations/test", f"{base_path}/test.csv")
````

````python
import os
import pandas as pd
import numpy as np
from PIL import Image
import tifffile
from torch.utils.data import Dataset
import torchvision.transforms as transforms
import ast

class F3CSVMultiLabelDataset(Dataset):
    def __init__(self, csv_file, transform=None):
        self.data = pd.read_csv(csv_file)
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data.iloc[idx, 0]
        label_str = self.data.iloc[idx, 1]
        label_list = ast.literal_eval(label_str)
        label = torch.tensor(label_list, dtype=torch.float32)

        image_np = tifffile.imread(img_path)
        if image_np.dtype != np.uint8:
            image_np = 255 * (image_np - image_np.min()) / (image_np.max() - image_np.min())
            image_np = image_np.astype(np.uint8)

        if image_np.ndim == 2:
            image_np = np.stack([image_np] * 3, axis=-1)

        image = Image.fromarray(image_np)

        if self.transform:
            image = self.transform(image)

        return image, label

class ResNetMultiLabel(nn.Module):
    def __init__(self, num_classes=6):
        super(ResNetMultiLabel, self).__init__()
        self.backbone = models.resnet18(weights=None)
        in_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.backbone(x)


def get_data_loaders(base_path, batch_size=16):
    transform_pipeline = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    train_dataset = F3CSVMultiLabelDataset(os.path.join(base_path, 'train.csv'), transform_pipeline)
    val_dataset = F3CSVMultiLabelDataset(os.path.join(base_path, 'val.csv'), transform_pipeline)
    test_dataset = F3CSVMultiLabelDataset(os.path.join(base_path, 'test.csv'), transform_pipeline)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader, test_loader
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
import numpy as np
from sklearn.metrics import f1_score, classification_report, accuracy_score

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=15, device="cuda", patience=5):
    model.to(device)
    history = {"train_loss": [], "val_loss": [], "val_f1": []}

    best_val_loss = float("inf")
    epochs_without_improvement = 0
    best_model_state = None

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        history["train_loss"].append(epoch_loss)

        model.eval()
        val_loss = 0.0
        all_probs = []
        all_labels = []

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * images.size(0)

                probs = torch.sigmoid(outputs).cpu().numpy()
                all_probs.append(probs)
                all_labels.append(labels.cpu().numpy())

        val_loss /= len(val_loader.dataset)
        all_probs = np.vstack(all_probs)
        all_labels = np.vstack(all_labels)

        preds = (all_probs > 0.5).astype(int)
        val_f1 = f1_score(all_labels, preds, average='macro')

        history["val_loss"].append(val_loss)
        history["val_f1"].append(val_f1)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {epoch_loss:.4f} - Val Loss: {val_loss:.4f} - Val F1: {val_f1:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_without_improvement = 0
            best_model_state = model.state_dict()
        else:
            epochs_without_improvement += 1
            if epochs_without_improvement >= patience:
                print(f"Early stopping at epoch {epoch+1}")
                break

    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    return history

def find_best_thresholds(y_true, y_probs):
    thresholds = []
    for i in range(y_true.shape[1]):
        best_thresh = 0.5
        best_f1 = 0
        for t in np.arange(0.1, 0.9, 0.05):
            preds = (y_probs[:, i] > t).astype(int)
            f1 = f1_score(y_true[:, i], preds, zero_division=0)
            if f1 > best_f1:
                best_f1 = f1
                best_thresh = t
        thresholds.append(best_thresh)
    return np.array(thresholds)


def evaluate_model(model, test_loader, device="cuda", thresholds=None):
    model.to(device)
    model.eval()

    all_probs = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = torch.sigmoid(model(images)).cpu().numpy()
            all_probs.append(outputs)
            all_labels.append(labels.numpy())

    all_probs = np.vstack(all_probs)
    all_labels = np.vstack(all_labels)

    if thresholds is None:
        thresholds = np.array([0.5] * all_labels.shape[1])

    preds = (all_probs > thresholds).astype(int)

    print("Classification Report:\n")
    print(classification_report(all_labels, preds, target_names=[f"Class {i}" for i in range(all_labels.shape[1])]))
    print(f"Accuracy\n\n: {accuracy_score(all_labels, preds)}")


train_loader, val_loader, test_loader = get_data_loaders(base_path)
model = ResNetMultiLabel(num_classes=6)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
history = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=15, device=device)

model.eval()
all_probs, all_labels = [], []
with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        outputs = torch.sigmoid(model(images)).cpu().numpy()
        all_probs.append(outputs)
        all_labels.append(labels.numpy())

val_probs = np.vstack(all_probs)
val_labels = np.vstack(all_labels)
best_thresholds = find_best_thresholds(val_labels, val_probs)

evaluate_model(model, test_loader, device=device, thresholds=best_thresholds)
torch.save(model.state_dict(), "/content/drive/My Drive/MC886/assignment_3/data/f3/resultados/batch15/val_multilabel.pth")
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
from sklearn.metrics import multilabel_confusion_matrix


class CustomCNNMultiLabel(nn.Module):
    def __init__(self,
                 input_channels=3,
                 num_classes=6,
                 conv_channels=[32, 64, 128],
                 kernel_size=3,
                 use_batchnorm=True,
                 use_dropout=True,
                 dropout_prob=0.3,
                 use_maxpool=True):
        super(CustomCNNMultiLabel, self).__init__()

        self.feature_extractor = nn.Sequential()
        in_channels = input_channels

        for i, out_channels in enumerate(conv_channels):
            self.feature_extractor.add_module(f"conv{i}", nn.Conv2d(in_channels, out_channels, kernel_size, padding=1))

            if use_batchnorm:
                self.feature_extractor.add_module(f"bn{i}", nn.BatchNorm2d(out_channels))

            self.feature_extractor.add_module(f"relu{i}", nn.ReLU(inplace=True))

            if use_maxpool:
                self.feature_extractor.add_module(f"pool{i}", nn.MaxPool2d(kernel_size=2, stride=2))

            if use_dropout:
                self.feature_extractor.add_module(f"dropout{i}", nn.Dropout2d(p=dropout_prob))

            in_channels = out_channels

        dummy_input = torch.zeros(1, input_channels, 224, 224)
        with torch.no_grad():
            dummy_output = self.feature_extractor(dummy_input)
            self.flatten_dim = dummy_output.view(1, -1).shape[1]

        self.classifier = nn.Sequential(
            nn.Linear(self.flatten_dim, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.feature_extractor(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=15, device="cuda"):
    model.to(device)
    history = {"train_loss": [], "val_loss": [], "val_accuracy": []}

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        history["train_loss"].append(epoch_loss)

        model.eval()
        val_loss = 0.0
        total = 0
        correct = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = torch.sigmoid(model(images))
                loss = criterion(outputs, labels)
                val_loss += loss.item() * images.size(0)

                preds = (outputs > 0.5).float()
                correct += (preds == labels).sum().item()
                total += labels.numel()

        val_loss /= len(val_loader.dataset)
        accuracy = correct / total
        history["val_loss"].append(val_loss)
        history["val_accuracy"].append(accuracy)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {epoch_loss:.4f} - Val Loss: {val_loss:.4f} - Val Acc: {accuracy:.4f}")

    return history

def plot_multilabel_confusion_matrix(y_true, y_pred, class_names=None):
    cm = multilabel_confusion_matrix(y_true, y_pred)
    num_classes = cm.shape[0]

    fig, axes = plt.subplots(1, num_classes, figsize=(4 * num_classes, 4))
    if num_classes == 1:
        axes = [axes]

    for i in range(num_classes):
        ax = axes[i]
        conf = cm[i]
        im = ax.imshow(conf, interpolation='nearest', cmap=plt.cm.Blues)
        ax.set_title(f"Classe {i}" if class_names is None else class_names[i])
        tick_marks = np.arange(2)
        ax.set_xticks(tick_marks)
        ax.set_yticks(tick_marks)
        ax.set_xticklabels(['Neg', 'Pos'])
        ax.set_yticklabels(['Neg', 'Pos'])

        for j in range(2):
            for k in range(2):
                ax.text(k, j, format(conf[j, k], 'd'),
                        ha="center", va="center",
                        color="white" if conf[j, k] > conf.max() / 2 else "black")

        ax.set_ylabel('Real')
        ax.set_xlabel('Predito')

    plt.tight_layout()
    plt.show()

def evaluate_model(model, test_loader, device="cuda"):
    model.to(device)
    model.eval()

    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = torch.sigmoid(model(images)).cpu()
            preds = (outputs > 0.5).int()

            all_preds.extend(preds.numpy())
            all_labels.extend(labels.numpy())

    print("Classification Report:\n", classification_report(all_labels, all_preds, target_names=[f"Class {i}" for i in range(6)]))

    class_names = [f"Class {i}" for i in range(all_preds[0].shape[0])]
    plot_multilabel_confusion_matrix(all_labels, all_preds, class_names)



train_loader, val_loader, test_loader = get_data_loaders(base_path)
criterion = nn.BCEWithLogitsLoss()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
````

````python
import cv2 as cv
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from torchvision.transforms.functional import normalize, to_pil_image

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model.eval()
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None
        target_layer.register_forward_hook(self.save_activation)
        target_layer.register_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activations = output.detach()

    def save_gradient(self, module, grad_input, grad_output):
        self.gradients = grad_output[0].detach()

    def generate(self, input_tensor, class_idx):
        self.model.zero_grad()
        output = self.model(input_tensor)
        prob = torch.sigmoid(output)[0, class_idx]
        prob.backward()

        pooled_grads = torch.mean(self.gradients, dim=[0, 2, 3])
        activations = self.activations.squeeze(0)

        for i in range(activations.shape[0]):
            activations[i, :, :] *= pooled_grads[i]

        heatmap = torch.mean(activations, dim=0).cpu()
        heatmap = F.relu(heatmap)
        heatmap /= torch.max(heatmap)
        return heatmap.numpy()


def show_gradcam_on_image(img_tensor, heatmap, alpha=0.4):
    img = img_tensor.squeeze(0).cpu().numpy()
    img = np.transpose(img, (1, 2, 0))
    img = (img - img.min()) / (img.max() - img.min())
    heatmap = cv.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cv.applyColorMap(heatmap, cv.COLORMAP_JET)
    overlay = heatmap_color / 255.0 * alpha + img * (1 - alpha)

    plt.imshow(overlay)
    plt.axis("off")
    plt.show()

train_loader, val_loader, test_loader = get_data_loaders(base_path)

model = ResNetMultiLabel(num_classes=6).to(device)
model.load_state_dict(torch.load('/content/drive/My Drive/MC886/assignment_3/data/f3/resultados/batch15/val_multilabel.pth'))
model.eval()
gradcam = GradCAM(model, target_layer=model.backbone.layer4)

for images, labels in test_loader:
    image = images[0].unsqueeze(0).to(device)
    true_labels = labels[0]

    outputs = torch.sigmoid(model(image))
    predicted_labels = (outputs > 0.5).int()[0]

    for class_idx in range(predicted_labels.shape[0]):
        if predicted_labels[class_idx] == 1:
            print(f"Visualizing Grad-CAM for Class {class_idx} (Predicted as present)")
            heatmap = gradcam.generate(image, class_idx)
            show_gradcam_on_image(image.cpu(), heatmap)

    break


````
